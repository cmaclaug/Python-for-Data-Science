```python
import pandas as pd
```

```python
world = pd.read_csv('Tennis_World_Number_1s.csv')
```

```python
world
```

```python
world.drop('Unnamed: 7',axis=1,inplace=True)
```

```python
world.columns
```

```python
world.shape
```

```python
world.isna().sum()
```

```python
world.dtypes
```

```python
world[['Career_Wins','Career_Losses']]=world.Careers_Wins_Losses.str.split('-',expand=True).astype(float)
```

```python
world.dtypes
```

```python
world[['Tennis_Player','Career_Wins','Career_Losses','Careers_Wins_Losses']].head()
```

```python
world['Total_Professional_Matches']=world.Career_Wins + world.Career_Losses
world['Total_Win_Percentage']=(world.Career_Wins / world.Total_Professional_Matches) * 100
world[['Tennis_Player','Total_Professional_Matches','Total_Win_Percentage']].sort_values('Total_Win_Percentage',ascending=False).head(10)
```

```python
world[['Tennis_Player','Years_End_no_1','Total_Win_Percentage']].sort_values(['Years_End_no_1','Total_Win_Percentage'],ascending=False).head(10)
```

```python
world.sort_values(['Prize_Money'],ascending=False)
```

```python
Tennis_Win_Perc = world.sort_values('Total_Win_Percentage', ascending = False)
Tennis_Win_Perc['Percent_Diff']=Tennis_Win_Perc.Total_Win_Percentage.pct_change()*100
Tennis_Win_Perc['Percent_Diff']=Tennis_Win_Perc.Percent_Diff.fillna(0)
Tennis_Win_Perc[['Tennis_Player','Total_Win_Percentage','Percent_Diff']].head(10)
```
